Time of the Day 

```python
fromDateFeature.head(3)
```

```python
def timeOfDay(t):
    if int(t)>=3 and int(t)<=7:
        return "Early Morning"
    elif int(t)>7 and int(t)<=12:
        return "Morning"
    elif int(t)>12 and int(t)<=16:
        return "Afternoon"
    elif int(t)>16 and int(t)<=20:
        return "Evening"
    elif int(t)>20 and int(t)<=23:
        return "Night"
    elif int(t)>=23 :
        return "MidNight"
    elif int(t)<=3 :
        return "MidNight"
```

```python
for hour in fromDateFeature['fromDateFeature_timestamp'].head():
    print(hour.split(":"))
    print(hour.split(":")[0])
```

```python
L = list()
for hour in fromDateFeature['fromDateFeature_timestamp']:
    L.append(timeOfDay(hour.split(":")[0]))
fromDateFeature['timeOfDay'] = L
```

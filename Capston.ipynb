Time of the Day 

```python
fromDateFeature.head(3)
```

```python
def timeOfDay(t):
    if int(t)>=4 and int(t)<=8:
        return "Early Morning"
    elif int(t)>8 and int(t)<=12:
        return "Morning"
    elif int(t)>12 and int(t)<=16:
        return "Afternoon"
    elif int(t)>16 and int(t)<=20:
        return "Evening"
    elif int(t)>20 and int(t)<=23:
        return "Night"
    elif int(t)>=23 :
        return "MidNight"
    elif int(t)<=4 :
        return "MidNight"
```

```python
for hour in fromDateFeature['fromDateFeature_timestamp'].head():
    print(hour.split(":")[0])
```
